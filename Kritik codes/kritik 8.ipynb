```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def gradient_descent(x0, y0, grad_f, alpha, num_iterations):
    x, y = x0, y0  # initialize x and y with the initial point

    for i in range(num_iterations):
        grad_x, grad_y = x - alpha * grad_x, y - alpha * grad_y #gradient descent
        return x, y

def fun_1(x, y): #function 1
    return x**2 + y**2

def grad_f_1(x, y): #derivative of f1
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

result_1 = gradient_descent(0.1, 0.1, grad_f_1, 0.1, 10) #result using given parameters
print("Result for fun_1:", result_1)

result_2 = gradient_descent(-1, 1, grad_f_1, 0.01, 100) #result using given parameters
print("Result for fun_1:", result_2)

def fun_2(x, y): #function 2
    return 1 - np.exp(-x**-2) * (y - 2)**2 - 2 * np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y): #derivative of f2
    epsilon = 1e-10 #needed this or the code would not work :( prevents division by 0 which does not work 
    grad_x = 2 * x * np.exp(-1 / (x**2 + epsilon)) * (y - 2)**2 + 4 * x * np.exp(-x**2 - (y + 2)**2)
    grad_y = -2 * np.exp(-1 / (x**2 + epsilon)) * (y - 2)**2 + 4 * (y + 2) * np.exp(-x**2 - (y + 2)**2)
    return grad_x, grad_y

result_3 = gradient_descent(0, 1, grad_f_2, 0.01, 10000)
print("Result for fun_2:", result_3)

result_4 = gradient_descent(0, -1, grad_f_2, 0.01, 10000)
print("Result for fun_2:", result_4)


#given plotting code
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(X, Y)
z = fun_2(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

```
