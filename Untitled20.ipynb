```python
import sympy as sp


#define symbols
x, y = sp.symbols('x y')

#def function
f = sp.exp(x)*(sp.sin(y)) + y**3

#compute partials
df_dx = sp.diff(f, x)
df_dy=sp.diff(f,y)

print ("The partial derivative with respect to x is", df_dx, "and the partial derivative with respect to y is", df_dy,".")
```

```python
#define symbols
x, y = sp.symbols('x y')

#define function
g =x**2*y + x*y**2
#define partials
df_dx = sp.diff(g, x)
df_dy=sp.diff(g,y)

# evaluate partials at (1, -1)
x_value = 1
y_value = -1
df_dx_at_point = df_dx.subs({x: x_value, y: y_value})
df_dy_at_point = df_dy.subs({x: x_value, y: y_value})

# Construct the gradient vector
gradient_vector = (df_dx_at_point, df_dy_at_point)

#magnitude formula

magnitude=sp.sqrt ((df_dx_at_point**2)+(df_dy_at_point**2))

print("Gradient vector at (1, -1):", gradient_vector)
print ("Magnitude of vector at (1,-1);",magnitude)
```

```python
#define symbols
x, y = sp.symbols('x y')

#define function
h=sp.log(x**2+y**2)
#define partials
df_dx = sp.diff(h, x)
df_dy=sp.diff(h,y)
#second partials
df_dx2=sp.diff(df_dx,x)
df_dy2=sp.diff(df_dy,y)
df_dxdy=sp.diff(df_dy,x)

print("Partial 1 with respect to x:",df_dx)
print("Partial 1 with respect to y:",df_dy)
print("Partial 2 with respect to x:",df_dx2)
print("Partial 2 with respect to y:",df_dy2)
print("Mixed partial:",df_dxdy)
```

```python
#import necessary libraries
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#defining the function
x,y=sp.symbols('x y')
j=x**3 - 3*x*y + y**3

#converting expression to numerical function
j_func=sp.lambdify((x,y),j,'numpy')

#creating a grid and evaluating j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#plot contour
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#import necessary libraries
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#defining the function
x,y=sp.symbols('x y')
j= sp.sqrt(x**2+y**2+x*y)

#converting expression to numerical function
j_func=sp.lambdify((x,y),j,'numpy')

#creating a grid and evaluating j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#plot contour
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = \sqrt{x^2 - y^2 + xy}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
